```python
def test_stat(stock,start,end,sma):
  df = yf.download(stock,start=start,end=end)
  data = df.Close.to_frame()
  data["returns"] = np.log(data.Close.div(data.Close.shift(1)))
  data["sma_s"] = data.Close.rolling(int(sma[0])).mean()
  data["sma_l"] = data.Close.rolling(int(sma[1])).mean()
  data.dropna(inplace = True)
  data["pos"] = np.where(data["sma_s"]>data["sma_l"],1,-1)
  data["stat"] = data["returns"]*data.pos.shift(1)
  data.dropna(inplace = True)
  ret = np.exp(data["stat"].sum())
  std = data["stat"].std()*np.sqrt(252)
  return ret,std
```

```python
from os import stat_result
class smabacktest():
  def __init__(self,symbol,sma_s,sma_l,start,end):
    self.symbol = symbol
    self.sma_s = sma_s
    self.sma_l = sma_l
    self.end = end
    self.start = start
    self.results = None
    self.get_data()

  def get_data(self):
    df = yf.download(self.symbol,start=self.start,end=self.end)
    data = df.Close.to_frame()
    data["returns"] = np.log(data.Close.div(data.Close.shift(1)))
    data["sma_s"] = data.Close.rolling(int(self.sma_s)).mean()
    data["sma_l"] = data.Close.rolling(int(self.sma_l)).mean()
    data.dropna(inplace = True)
    self.data2 = data
    return data

  def test_result(self):
    data = self.data2.copy().dropna()
    data["pos"] = np.where(data["sma_s"]>data["sma_l"],1,-1)
    data["stat"] = data["returns"]*data.pos.shift(1)
    data.dropna(inplace = True)
    data["retbh"] = data["returns"].cumsum().apply(np.exp)
    data["stat_result"] = data["stat"].cumsum().apply(np.exp)
    perf = data["stat_result"].iloc[-1]
    outperf = perf-data["retbh"].iloc[-1]
    self.results = data

    ret = np.exp(data["stat"].sum())
    std = data["stat"].std()*np.sqrt(252)
    return round(perf,6),round(outperf,6)
  def plot_result(self):
    if self.results is None:
      print("run the test please")
    else:
      title="{}\sma_s{} | sma_l{}".format(self.symbol,self.sma_s,self.sma_l)
      self.results[["retbh","stat_result"]].plot(title=title ,figsize = (12,8))



```
